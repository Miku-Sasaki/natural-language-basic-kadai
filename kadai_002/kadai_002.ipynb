```python
import requests
from bs4 import BeautifulSoup

# ① Beautiful Soup4を用いてスクレイピング
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"

# リクエストを送信
response = requests.get(url)

# 青空文庫の文字化けを防ぐため、適切なエンコーディングを設定
response.encoding = response.apparent_encoding

# ステータスコードが200（成功）か確認
print(f"ステータスコード: {response.status_code} (200なら成功です)")

if response.status_code == 200:
    print("解析に成功しました。データの読み込みを行っています...\n")

    # HTMLをパース
    soup = BeautifulSoup(response.text, "html.parser")

    # 本文（main_text）の部分を取得
    main_text_element = soup.find("div", class_="main_text")

    if main_text_element:
        # 最初の200文字だけ試しに表示して、正しく取得できているか確認
        print("--- 取得したテキストの冒頭部分 ---")
        print(main_text_element.get_text()[:200])
        print("--------------------------------")
    else:
        print("main_textタグが見つかりませんでした。")
else:
    print("URLの読み込みに失敗しました。URLを確認してください。")
```

```python
import re

# ステップ①で `main_text_element` が無事に取得できている前提で進めます
if "main_text_element" in locals() and main_text_element:
    # 1. HTMLタグを取り除き、純粋なテキストのみを抽出
    # (Beautiful Soupの get_text() を使うことで、HTMLタグ自体は自動で削除されます)
    raw_text = main_text_element.get_text()

    # 2. 改行や空白のクレンジング処理
    # 文字列の前後にある余分な空白や改行を削除
    cleaned_text = raw_text.strip()

    # 連続する複数の改行（\n\nなど）を、1つの改行にまとめます
    cleaned_text = re.sub(r"\n+", "\n", cleaned_text)

    # 3. 結果の確認（最初の5行と最後の5行を表示してみます）
    lines = cleaned_text.splitlines()

    print("--- クレンジング後の本文（最初の5行） ---")
    for line in lines[:5]:
        print(line)

    print("\n----------------------------------------\n")

    print("--- クレンジング後の本文（最後の5行） ---")
    for line in lines[-5:]:
        print(line)
    print("----------------------------------------")

else:
    print(
        "ステップ①のテキストデータ（main_text_element）が見つかりません。先に取り込みを実行してください。"
    )
```

```python
# ③ 最初の1文（正しい文章）を手動で品詞分解してストップワードを除去

# 正しい最初の1文
first_sentence = "近頃は大分方々の雑誌から談話をしろしろと責められて、頭ががらん胴になったから、当分品切れの看板でも懸けたいくらいに思っています。"
print(f"【対象の文】\n{first_sentence}\n")

# 1. 正しい文章に基づく手動での品詞分解（単語, 品詞）
manual_tokens = [
    ("近頃", "名詞"),
    ("は", "助詞"),
    ("大分", "副詞"),
    ("方々", "名詞"),
    ("の", "助詞"),
    ("雑誌", "名詞"),
    ("から", "助詞"),
    ("談話", "名詞"),
    ("を", "助詞"),
    ("しろしろ", "副詞"),  # 擬態語・副詞
    ("と", "助詞"),
    ("責め", "動詞"),
    ("られ", "助動詞"),
    ("て", "助詞"),
    ("、", "記号"),
    ("頭", "名詞"),
    ("が", "助詞"),
    ("がらん胴", "名詞"),  # 「がらんどう」
    ("に", "助詞"),
    ("なっ", "動詞"),
    ("た", "助動詞"),
    ("から", "助詞"),
    ("、", "記号"),
    ("当分", "名詞"),  # または副詞
    ("品切れ", "名詞"),
    ("の", "助詞"),
    ("看板", "名詞"),
    ("で", "助詞"),
    ("も", "助詞"),
    ("懸け", "動詞"),
    ("たい", "助動詞"),
    ("くらい", "助詞"),  # 副助詞
    ("に", "助詞"),
    ("思っ", "動詞"),
    ("て", "助詞"),
    ("い", "動詞"),
    ("ます", "助動詞"),
    ("。", "記号"),
]

# 2. ストップワード（除去対象）の定義
# 助詞、助動詞、記号を除外します
stop_pos = ["助詞", "助動詞", "記号"]

# 3. ストップワードの除去処理
filtered_tokens = [
    (word, pos) for word, pos in manual_tokens if pos not in stop_pos
]

# 4. 結果の表示
print("--- ストップワード除去後の単語一覧 ---")
for word, pos in filtered_tokens:
    print(f"・{word} ({pos})")
```
